```python
import pandas as pd

# Read in the data
df = pd.read_excel('data/PrixElec.xlsx')
df.drop(columns=['Unnamed: 2','Unnamed: 5','Unnamed: 8' ], inplace=True)
cols = ['Samedi', 'Unnamed: 4', 'Dimanche', 'Unnamed: 7']
columns_to_drop = [col for col in df.columns if col not in cols]
df.drop(columns=columns_to_drop, inplace=True)
df.rename(columns={'Unnamed: 4': 'ct/KWh(Samedi)', 'Unnamed: 7': 'ct/KWh(Dimanche)'}, inplace=True)
df = df.drop(df.index[0])
df['Samedi'] = pd.to_datetime(df['Samedi'], format='%H', errors='coerce')
# Set a specific date '1.12.23' for the 'Samedi' column while maintaining the hour
df['Samedi'] = df['Samedi'].apply(lambda x: x.replace(year=2023, month=12, day=1) if pd.notnull(x) else pd.NaT)

df['Dimanche'] = pd.to_datetime(df['Samedi'], format='%H', errors='coerce')
# Set a specific date '1.12.23' for the 'Samedi' column while maintaining the hour
df['Dimanche'] = df['Samedi'].apply(lambda x: x.replace(year=2023, month=12, day=2) if pd.notnull(x) else pd.NaT)

df
```

```python
result = pd.concat([df['Samedi'], df['ct/KWh(Samedi)']], axis=1)
result.columns = ['Datetime', 'ct/KWh']
result2= pd.concat([df['Dimanche'], df['ct/KWh(Dimanche)']], axis =1) 
result2.columns = ['Datetime', 'ct/KWh']
result
result2
```

```python
merged_df = pd.concat([result, result2], ignore_index=True)
merged_df
merged_df.to_csv('data/electricity_price.csv', index=False)
```
