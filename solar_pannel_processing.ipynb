```python
import pandas as pd

df = pd.read_csv('data/database_test_PV_Power.csv', sep=';')

df
```

```python
df.columns = ['DateTime', 'Nan', 'Power']
```

```python
start_date = pd.Timestamp('2023-12-01 00:00:00')
end_date = pd.Timestamp('2023-12-03 00:00:00')
hourly_range = pd.date_range(start=start_date, end=end_date, freq='H')
new_df = pd.DataFrame(hourly_range, columns=['DateTime'])
```

```python
df['DateTime'] = new_df['DateTime']
df.drop('Nan', axis=1, inplace=True)
df['Power'] = df['Power'].str.replace(',', '.').astype(float)
df
file_path = 'data/df_PV.csv'  # Replace 'output.csv' with your desired file path
df.to_csv(file_path, index=False) 
```
